```python
import pandas as pd
import numpy as np
```

```python
a = pd.read_csv("/content/drive/MyDrive/ML Datasets/car data.csv")
df = pd.DataFrame(a)
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns="Car_Name")
```

```python
df["Fuel_Type"].value_counts()
```

```python
df["Fuel_Type"] = df["Fuel_Type"].replace("Petrol", 1)
df["Fuel_Type"] = df["Fuel_Type"].replace("Diesel",2)
df["Fuel_Type"] = df["Fuel_Type"].replace("CNG", 3)
```

```python
df["Fuel_Type"].value_counts()
```

```python
df["Seller_Type"].value_counts()
```

```python
df["Seller_Type"] = df["Seller_Type"].replace("Dealer", 1)
df["Seller_Type"] = df["Seller_Type"].replace("Individual", 0)
```

```python
df["Transmission"].value_counts()
```

```python
df["Transmission"] = df["Transmission"].replace("Manual", 1)
df["Transmission"] = df["Transmission"].replace("Automatic",0)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
X = df.drop(columns="Selling_Price").values
```

```python
Y = df["Selling_Price"].values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2, shuffle=True)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train.std()
```

```python
X_test.std()
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train, Y_train)
```

```python
from sklearn import metrics
```

```python
X_train_prediction = model.predict(X_train)
```

```python
r_squared_training = metrics.r2_score(X_train_prediction, Y_train)
print(r_squared_training)
```

```python
X_test_prediction = model.predict(X_test)
```

```python
r_squared_training = metrics.r2_score(X_test_prediction, Y_test)
print(r_squared_training)
```

```python
import gradio as gr
import joblib
import numpy as np

# Load the model and scaler
model = joblib.load("car_price.pkl")
scaler = joblib.load("scaler.pkl")

# Prediction function
def predict_price(year, present_price, kms_driven, fuel_type, seller_type,
                  transmission, owners):

    # Convert categories to numerical labels
    fuel_map = {"Petrol": 1, "Diesel": 2, "CNG": 3}
    seller_map = {"Dealer": 1, "Individual": 0}
    transmission_map = {"Manual": 1, "Automatic": 0}

    fuel_type = fuel_map[fuel_type]
    seller_type = seller_map[seller_type]
    transmission = transmission_map[transmission]

    # Create input array
    input_data = np.array([[year, present_price, kms_driven, fuel_type,
                            seller_type, transmission, owners]])

    # Scale the input
    scaled_input = scaler.transform(input_data)

    # Predict using the model
    prediction = model.predict(scaled_input)[0]
    return f"ðŸš— Estimated Selling Price: â‚¹ {prediction:.2f} lakhs"

# Gradio interface
interface = gr.Interface(
    fn=predict_price,
    inputs=[
        gr.Number(label="Year"),
        gr.Number(label="Present Price (in lakhs)"),
        gr.Number(label="Kms Driven"),
        gr.Radio(["Petrol", "Diesel", "CNG"], label="Fuel Type"),
        gr.Radio(["Dealer", "Individual"], label="Seller Type"),
        gr.Radio(["Manual", "Automatic"], label="Transmission"),
        gr.Number(label="Number of Owners")
    ],
    outputs=gr.Text(label="Predicted Selling Price"),
    title="ðŸš˜ Car Price Predictor",
    description="Enter the details of your car to estimate its selling price using our trained Random Forest model."
)

interface.launch()
```
